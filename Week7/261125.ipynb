```python
import time
import random

print("Get ready.....")
wait_time = random.randint(2, 5)
time.sleep(wait_time)

print("GO!!!")
start_time =  time.time()

input("Press Enter as quickly as possible")

end_time = time.time()
reaction_time = end_time - start_time
print(f"Your reaction time: {reaction_time :.3f} seconds")

if reaction_time >=0.7:
    print("You have the reaction time of a donkey")

if reaction_time <=0.7>0.3:
    print("You have a good reaction time")

if reaction_time <=0.3:
    print("WOAH, you have the reaction time of a F1 Driver")
    
```

```python
def volumeCube():
    while True:
        try:
            length = int(input("length:"))
            width = int(input("width:"))
            height = int(input("height:"))

            volume = length*height*width
            print(f'Volume is {volume} cubic units')
            break
        except ValueError:
            print('only numbers may be entered')
        
volumeCube()

```

```python
def main():
    menu = '''
 1. Volume of a cube
 2. Pythagoras Theorem
 3. Percentage Change
 4. Quit'''
    while True:
        print(menu)
        while True:
            try:
                menuChoice = int(input('Enter a menu choice:'))
                break
            except:
                print("Please enter a number from the menu:")
        if menuChoice == 4:
            break 
        elif menuChoice ==1:
            volumeCube()
main()
```

```python
import math 
def Pythagoras():
    while True:
        try:
            adjacent = int(input('adjacent length:'))
            opposite = int(input('opposite length:'))

            hypotenuse = math.sqrt((adjacent*adjacent) + (opposite*opposite))
            print(f'Hypotenuse length is {hypotenuse} cubic units')
            break
        except ValueError:
            print('only numbers may be entered')
Pythagoras()
```
